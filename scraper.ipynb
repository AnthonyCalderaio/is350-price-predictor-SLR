```python
import requests
from bs4 import BeautifulSoup
import xlsxwriter
import re
import os
```

```python
# Specify the URL of the car website you want to scrape
urlList = [
    # "https://www.cars.com/shopping/results/?makes[]=lexus&maximum_distance=30&models[]=lexus-is_350&page=1&stock_type=all&zip=33418",
    # "https://www.cars.com/shopping/results/?makes[]=lexus&maximum_distance=30&models[]=lexus-is_350&page=2&stock_type=all&zip=33418"
    "https://www.cars.com/shopping/results/?dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=lexus&maximum_distance=75&mileage_max=&models[]=lexus-is_350&monthly_payment=&page=1&page_size=20&sort=best_match_desc&stock_type=all&year_max=&year_min=&zip=33418",
    "https://www.cars.com/shopping/results/?dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=lexus&maximum_distance=75&mileage_max=&models[]=lexus-is_350&monthly_payment=&page=2&page_size=20&sort=best_match_desc&stock_type=all&year_max=&year_min=&zip=33418",
    "https://www.cars.com/shopping/results/?dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=lexus&maximum_distance=75&mileage_max=&models[]=lexus-is_350&monthly_payment=&page=3&page_size=20&sort=best_match_desc&stock_type=all&year_max=&year_min=&zip=33418",
    "https://www.cars.com/shopping/results/?dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=lexus&maximum_distance=75&mileage_max=&models[]=lexus-is_350&monthly_payment=&page=4&page_size=20&sort=best_match_desc&stock_type=all&year_max=&year_min=&zip=33418"

]
url = "https://www.cars.com/shopping/results/?stock_type=all&makes%5B%5D=lexus&models%5B%5D=lexus-is_350&maximum_distance=30&zip=33418#vehicle-card-fe28e232-8c0c-4c15-aa15-3ed20041d597"

# Remove excel sheet to replace
try:
    os.remove("is300_cars.xlsx")
except:
    print("Nothing to replace. Proceeding.")

# Excel sheet
workbook = xlsxwriter.Workbook('is300_cars.xlsx')
worksheet = workbook.add_worksheet()
row = 1
column = 0

worksheet.write("A1", "Cost")
worksheet.write("B1", "Mileage")
worksheet.write("C1", "Year")

for url in urlList:

    # Send a GET request to the URL and retrieve the content
    response = requests.get(url)
    html_content = response.content

    # Parse the HTML content with Beautiful Soup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the elements containing the information you need (modify based on website structure)
    car_listings = soup.find_all('div', class_='vehicle-card-main')  # Example class, adjust as per your website

    # Iterate through each car listing and extract details
    for car in car_listings:
        car_entry = []
        price = ''
        year = ''
        mileage = ''
        price = re.sub("[^\d]", "",car.find('span', class_='primary-price').text.strip())
        mileage_element = car.find('div', class_='vehicle-details').find('div', class_='mileage')
        if mileage_element is not None:
            mileage = re.sub("[^\d]", "",mileage_element.text.strip().strip())
        else:
            mileage = '';
        # vehicle-details car.find('div', class_='vehicle-details').find('div', class_='mileage').text.strip()
        # color = 'test'#car.find('span', class_='color').text.strip()
        year = car.find('a', class_='vehicle-card-link').text.strip()[:4]
        car_entry = [price, mileage, year]
        column = 0

        # If all the data exists, write to it
        if(price and mileage and year):
            if(len(car_entry) is 3):
                for item in car_entry:
                    
                    print('added: '+str(car_entry) + ' at row:'+str(row))
                    worksheet.write(row, column, item)
                    # Add column
                    column += 1
            car_entry = []
            row += 1


# Write to excel sheet
workbook.close()

```
